## inheritance

```
# ParentClass
class Employee:
  def __init__(self, name, base_salary):
    self.name = name
    self.base_salary = base_salary

  def calculate_pay(self):
    return self.base_salary

# ChildClass (Manager)
class Manager(Employee):
  def __init__(self , name, base_salary, bonus):
    super().__init__(name, base_salary)
    self.bonus = bonus

  def calculate_pay(self):
    return self.base_salary + self.bonus

# ChildClass (Developer) while the method action is defferent
class Developer(Employee):
  def __init__(self, name, base_salary, program_language):
    super().__init__(name, base_salary)
    self. program_language =  program_language

  def write_code(self):
    print(f"{self.name} is writing code in {self. program_language}")

print("\n--------Child Class Manager-------\n")

manager1 = Manager("Sokha", 1500, 300)

print(f"Manager: {manager1.name}")
print(f"Total Pay: ${manager1.calculate_pay()}")

print("\n--------Child Class Developer-------\n")

developer1 = Developer("Vath",1500, "Python")

print(f"Developer Name: {developer1.name}")
print(f"Salary: {developer1.base_salary}$")
print(f"Language: {developer1.program_language}")
```
